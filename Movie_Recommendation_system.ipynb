# IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# DATA COLLECTION AND PRE-PROCESSING

```python
# Loading the dataset

movies_data = pd.read_csv("movies.csv")
```

```python
# printing first five rows of the dataframe

movies_data.head()
```

```python
# Number of rows and columns in the dataframe

movies_data.shape
```

```python
# Selecting the relevant features for recommendation

selected_features = ["genres","keywords","original_language","original_title","popularity","tagline","cast","director"]
print(selected_features)
```

```python
movies_data.isnull().sum()
```

```python
# Replacing null values with null strings

for features in selected_features:
    movies_data[features] = movies_data[features].fillna('')
```

```python
# Combining all the selected features

movies_data['popularity'] = movies_data['popularity'].astype(str)

combine = movies_data["genres"]+' '+movies_data["keywords"]+' '+movies_data["original_language"]+' '+movies_data["original_title"]+' '+movies_data["popularity"]+' '+movies_data["tagline"]+' '+movies_data["cast"]+' '+' '+movies_data["director"]
```

```python
print(combine)
```

```python
# Converting text data into feature vector

vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combine)
```

```python
print(feature_vector)
```

# COSINE SIMILARITY

```python
# Getting the similarity score using cosine similarity

similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

# GETTING INPUT FROM USER

```python
# getting input
movie_name = input("Enter the moive name here :) ")
```

```python
# Creating a list with all the movies name given in the dataset
movies_titles = movies_data["title"].tolist()
print(movies_titles)
```

```python
# Finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,movies_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# Finding the index of the movie with title
index_of_movie = movies_data[movies_data.title == close_match]["index"].values[0]
print(index_of_movie)
```

```python
# Getting list of similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on index
print("movies suggested for you : \n")

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]["title"].values[0]
    if(i<=30):
        print(i,'.',title_from_index)
        i += 1
```
